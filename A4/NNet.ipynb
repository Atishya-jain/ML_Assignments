```python
Train_NN_X = torch.from_numpy(Train_NN_X)
Test_NN_X = torch.from_numpy(Test_NN_X)
Test_NN_Y = torch.from_numpy(Test_NN_Y)
Train_NN_Y = torch.from_numpy(Train_NN_Y)
```

```python
Train_NN_X = Train_NN_X.float().cuda()
Test_NN_X = Test_NN_X.float().cuda()
Test_NN_Y = Test_NN_Y.long().cuda()
Train_NN_Y = Train_NN_Y.long().cuda()
print(Train_NN_Y)
```

```python
n_in, n_h, n_out, batch_size = 6720, 1000, 2, 100
# model = nn.Sequential(nn.Linear(n_in, n_h),
#                      nn.Sigmoid(),
#                      nn.Linear(n_h, n_out),
#                      nn.LogSoftmax()).cuda()
model = nn.Sequential(nn.Linear(n_in, n_h),
                     nn.Sigmoid(),
                     nn.Linear(n_h, n_out)).cuda()
# criterion = torch.nn.NLLLoss().float()

train_set = my_dataloader('data/train')
train_generator = DataLoader(train_set, batch_size=batch_size, shuffle=True, num_workers=4)
# ={'batch_size': batch_size, 'shuffle': True, 'num_workers':4}
criterion = torch.nn.CrossEntropyLoss().float()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
# print(Train_NN_X.shape[0])
for epoch in range(50):
#     for i in range(int(Train_NN_X.shape[0]/batch_size) + 1):
    for x,y in train_generator:    
#         y_pred = model(Train_NN_X[i*batch_size:min((i+1)*batch_size,Train_NN_X.shape[0])])
        y_pred = model(x.cuda())
#         orig = Train_NN_Y[i*batch_size:min((i+1)*batch_size,Train_NN_Y.shape[0])]
#         print(y_pred)
#         print(orig)
#         print(1/0)
#         loss = criterion(y_pred, Variable(orig))
        loss = criterion(y_pred, y.cuda())
        print('epoch: ', epoch,' loss: ', loss.item())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
